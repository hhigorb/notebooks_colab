**Seleções e Frequências**

```
import pandas as pd
```

```
df = pd.read_csv('/content/drive/MyDrive/bases_de_dados/aluguel_residencial.csv', sep=';')
df.head(10)
```

**Atividades**

*Selecione somente os imóveis classificados com tipo 'Apartamento'.*



```
imovel_apartamento = df.query('Tipo == "Apartamento"')

# organizando os indices

def organiza_indice(dataset):
  dataset.index = range(len(dataset))
  return dataset

organiza_indice(imovel_apartamento)
```

*Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.*

```
# definindo os tipos solicitados
tipos_de_imoveis = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```
# filtrando somente imóveis que fazem parte do conjunto 'tipos_de_imoveis'
imoveis_tipo_casa = df.query('Tipo in @tipos_de_imoveis')

# organizando os indices

organiza_indice(imoveis_tipo_casa)
```

*Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.*

```
areas_60_a_100 = df.query('Area >= 60 and Area <= 100')

# organizando os indices

organiza_indice(areas_60_a_100)
```

*Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.*


```
imoveis_4_quartos_abaixo_2000 = df.query('Quartos >= 4 and Valor < 2000')

# organizando os indices

organiza_indice(imoveis_4_quartos_abaixo_2000)
```
