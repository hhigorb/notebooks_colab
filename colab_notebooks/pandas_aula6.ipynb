**Criando novas variáveis**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/bases_de_dados/aluguel_residencial2.csv', sep=';')
df.head(10)
```

```python
# criando a variável 'Valor Bruto'
df['Valor Bruto'] = df.loc[:, ['Valor', 'Condominio', 'IPTU']].sum(axis=1)
df
```

```python
# criando a variável 'Valor do metro quadrado'
df['Valor M²'] = df['Valor'] / df['Area']
df
```

```python
# arredondando os valores de Valor M²
df['Valor M²'] = df['Valor M²'].round(2) 
df
```

```python
# criando a variável Valor Bruto M²
df['Valor Bruto M²'] = (df['Valor Bruto'] / df['Area']).round(2)
df.head(10)
```

```python
# separando as variáveis Tipo
casas = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
df['Tipo Agregado'] = df['Tipo'].apply(lambda x: 'Casa' if x in casas else 'Apartamento')
df
```
