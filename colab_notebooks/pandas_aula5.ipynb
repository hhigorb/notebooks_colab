**Tratamento de Dados Faltantes**

```
import pandas as pd
```

```
dados = pd.read_csv('/content/drive/MyDrive/bases_de_dados/aluguel_residencial.csv', sep=';')
dados.head(10)
```

```
dados.isnull() # retorna um df com True caso a informação seja NaN
```

```
dados.notnull() # retorna False para informações nulas
```

```
dados.info()
```

```
# filtrando os valores Null da variável 'Valor'
dados.query('Valor == "NaN"')
```

```
# eliminando os valores Null da variável 'Valor'
dados.dropna(subset=['Valor'], inplace=True)
dados
```

```
# selecionando as variáveis apartamento que possuem condominio como NaN
selecao = dados.query('Tipo == "Apartamento" and Condominio == "NaN"')

dados.drop(selecao.index)
```

```
# substituindo o restante dos valores NaN para 0
dados.fillna(0, inplace=True)
dados
```

```
dados.info()
```

```
# exportando para csv
dados.to_csv('aluguel_residencial2.csv', sep=';', index=False)
```
