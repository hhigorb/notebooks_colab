```python
import pandas as pd
```

```python
dados = {
    'Nome': ['Higor', 'Marcelo', 'Ricardo', 'Márcio', 'Carla', 'Fernanda'],
    'Sobrenome': ['Henrique', 'Peixoto', 'Meireles', 'Cardoso', 'Teixeira', 'Cristina'],
    'Idade': [20, 22, 30, 40, 33, 22]
}

df = pd.DataFrame(dados)
df
```

```python
df['Nome Completo'] = df[['Nome', 'Sobrenome']].apply(lambda x: ' '.join(x), axis=1)
```
