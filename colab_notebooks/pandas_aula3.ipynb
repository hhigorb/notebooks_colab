# Relatório de Análise


**Imóveis Residenciais**


```
import pandas as pd
```

```
dados = pd.read_csv('/content/drive/MyDrive/bases_de_dados/aluguel.csv', sep=';')
dados.head()
```

```
# criando uma lista com a variável 'Tipo' do dataframe e removendo as duplicatas
list(dados['Tipo'].drop_duplicates())
```

```
# selecionei apenas os imóveis residenciais
residencial =  ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila']
residencial
```

```
# a função isin verifica se os valores passados por parâmetros estão contido no dataset, caso sim, retorna True
selecao = dados['Tipo'].isin(residencial)
selecao
```

```
# passando a seleção, será feita uma verificação booleana e retornará somente os imóveis residenciais
dados_residenciais = dados[selecao]
dados_residenciais.head(10)
```

```
# organizando os índices novamente
dados_residenciais.index = range(len(dados_residenciais))
dados_residenciais
```

# Exportando a base de dados

```
dados_residenciais.to_csv('aluguel_residencial.csv', sep=';', index=False) # index False pois se eu exportar como True quando for abrir o arquivo novamente irá abrir com os indices duplicados
```

**Organizar os dataframes com sort**

```
# sort_index -> organizar os indices
# sort_values -> organizar os valores do df


dados = [['Ary', 'M', 21], ['Cátia', 'F', 19], ['Carlos', 'M', '50'], ['Beto', 'M', 29], ['Bruna', 'F', 31], ['Ana', 'F', 42]]

df = pd.DataFrame(dados, columns=['Nome', 'Sexo', 'Idade'])
```

```
df
```

```
df2 = df.sort_values(['Sexo'])
df2
```

```
df2.sort_index(axis=1, inplace=True)
df2.index = range(len(df2))
df2
```
