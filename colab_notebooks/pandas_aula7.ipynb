**Criando Agrupamentos**

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/bases_de_dados/aluguel_residencial_atual.csv', sep=';')
dados.head()
```

```python
# criando uma média de aluguel baseado em cada bairro
grupo_de_bairros = dados.groupby('Bairro')
grupo_de_bairros[['Valor']].mean().round(2)
```

**Estatísticas Descritivas**

```python
grupo_de_bairros['Valor'].describe().round(2)
```

```python
grupo_de_bairros['Valor'].aggregate(['mean', 'count', 'sum'])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (10, 10))
```

```python
# fazendo o plot dos bairros
bairros_plot = grupo_de_bairros['Valor'].mean().plot.bar(color = 'red')
bairros_plot.set_ylabel('Valor do Aluguel')
bairros_plot.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
